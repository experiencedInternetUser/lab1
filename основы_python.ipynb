```python
def DoTests(func: callable, tests : list[list]) -> None:
    for index, test in enumerate(tests):
        if (func(test[1][0]) == test[1][1]):
            print(f"Test №{index + 1}({test[0]}) Successful")
            print("Input: " + str(test[1][0]))
            print("Output: " + str(func(test[1][0])))
            print()
        else:
            print(f"Test №{index + 1}({test[0]}) Failed")
            print("Input: " + str(test[1][0]))
            print("Your output: " + str(func(test[1][0])))
            print("Right output: " + str(test[1][1]))
            print()
            assert func(test[1][0]) == test[1][1]
```

Написать функцию на вход которой подается строка, состоящая из латинских букв.

```python
def foo_1(string: str) -> dict: 
    dictionary = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0,}
    for i in string:
        if (i in dictionary.keys()):
            dictionary[i] += 1
    return dictionary
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if (len(str(n)) == 1): return 0

    multiplications = 0
    while(True):
        previous = 1
        for i in str(n):
            result = previous * int(i)
            previous = result
        n = result
        
        multiplications += 1
        if (len(str(n)) == 1): 
            return multiplications
```

```python
tests = [
    ["One_Digit", [9, 0] ],
    ["Two_digit", [10, 1] ],
    ["Task_example", [39, 3] ],
    ["Task_example_2", [999, 4]],
    ["ten_thousands", [10_000, 1]],
]
         
DoTests(magic, tests)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(v1 : list[int], v2: list[int]) -> float:
    v1 = numpy.array(v1)
    v2 = numpy.array(v2)
    if (numpy.linalg.norm(v1) != numpy.linalg.norm(v2)):
        raise Exception("v1 length != v2 length")
    
    Mean_square_deviation = numpy.sqrt(numpy.mean((v1 - v2) ** 2))
    return Mean_square_deviation
```

```python
def test() -> None:
    funcResult = mse([[1], [2]], [[2], [1]])
    trueResult = 1
    assert funcResult == trueResult, f"Test №1 Failed  Your result: \n {str(funcResult)}"
    print("Test №1 Successful")

    funcResult = mse([10,100], [100,10])
    trueResult = 90
    assert funcResult == trueResult, f"Test №2 Failed  Your result: \n {str(funcResult)}"
    print("Test №2 Successful")

test()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def foo_6(number: int) -> int:
    if number < 0:
        raise ValueError
    
    prime_numbers = []
    for i in range(2, int(math.sqrt(number)) + 1): # обычно делитель не будет больше корня
        while (number % i == 0):
            prime_numbers.append(i)
            number /= i
    if (len(prime_numbers) == 0): return f"({number})"

    sorted_prime_numbers = sorted(set(prime_numbers))

    to_form = []
    for i in sorted_prime_numbers:
        count = prime_numbers.count(i)
        if (count != 1):
            string = f"({i}**{count})"
            to_form.append(string)
        else:
             to_form.append(f"({i})")
        
    return "".join(to_form)

print(foo_6(0))
```

```python
tests = [
    ["Zero", [0, "(0)"]],
    ["One", [1, "(1)"]],
    ["Common_Test", [4, "(2**2)"]],
    ["Common_Test", [54, "(2)(3**3)"]],
    ["Prime_number", [17, "(17)"]],
    ["Task_Example", [86240, "(2**5)(5)(7**2)(11)"]],
]           

DoTests(foo_6, tests)
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number: int) -> int | str:
    k = 1
    sum = 0
    while(number > sum):
        sum += k**2
        k += 1
    
    if (number == sum):
        return k - 1
    else:
        return "it is impossible"
    
print(pyramid(0))
```

```python
tests = [
    ["Zero", [0, 0]],
    ["One", [1, 1]],
    ["five", [5, 2]],
    ["fourteen", [14, 3]],
    ["thirty", [30, 4]],

    ["Could_Not", [-10, "it is impossible"]],
    ["Could_Not_2", [10, "it is impossible"]],
]           

DoTests(pyramid, tests)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def foo_8(number: int) -> bool:
    if (number < 0):
         raise ValueError
    
    strNumber = str(number)
    lengthNumber = len(strNumber)
    if (lengthNumber <= 2):
         raise ValueError
    
    ruptureSite = int(lengthNumber / 2 - 1 if lengthNumber % 2 == 0 else lengthNumber // 2)
    left= strNumber[:ruptureSite]
    right= strNumber[-ruptureSite:]

    sum_left, sum_right = 0, 0
    for i in range(len(left)):
          sum_left += int(left[i])
          sum_right += int(right[i])
    
    return sum_left == sum_right
```

```python
tests = [
    ["Common_Test", [22_99_04, True]],
    ["Common_Test", [11_99_55, False]],

    ["Common_Test", [222_9_006, True]],
    ["Common_Test", [11_989_55, False]],
]           

DoTests(foo_8, tests)
```
